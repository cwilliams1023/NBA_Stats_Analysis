```python
# we want to use the numerical data for the predictors (share of mvp votes)

predictors =['G', 'MP', 'PTS', 'TRB', 'AST', 'STL', 'BLK', 'FG%', '3P%', 'FT%']
```

```python
# train model based on years less than 2023

train = mvp[~(mvp['Year'] == 2023)]
```

```python
# make predictions on test year 2023

test = mvp[mvp['Year'] == 2023]
```

# **Identifying an error metric**

```python
#error metric helps you know if algo did well or not

from sklearn.metrics import mean_squared_error

mean_squared_error(combination['Share'], combination['predictions']) #mean difference between prediction and actual value
```

```python
# add rank column

combination = combination.sort_values('Share', ascending=False)
combination['Rk'] = list(range(1, combination.shape[0]+1))
combination
```

```python
#rank predictions

combination = combination.sort_values('predictions', ascending=False)
combination['Predicted_Rk'] = list(range(1, combination.shape[0]+1))
combination
```

```python
#create error metric

def find_ap(combination):
  actual = combination.sort_values('Share', ascending=False)
  predicted = combination.sort_values('predictions', ascending=False)
  ps=[]
  found = 0
  seen = 1
  for index, row in predicted.iterrows():
    if row['Player'] in actual['Player'].values:
      found += 1
      ps.append(found/seen)
    seen += 1
  return sum(ps) / len(ps)
```

```python
find_ap(combination)
```

# **Backtesting to predict the MVP each year**

```python
years = list(range(2015,2023))
```

```python
#start empty list to store precision values and predictions
aps = []
all_predictions = []
for year in years[3:]:
    train = mvp[mvp['Year'] < year]  #training set with data from years less than the current year
    test= mvp[mvp['Year'] == year] #create test set with data from the current year
    reg.fit(train[predictors], train['Share']) #fit regression model using training data
    predictions = reg.predict(test[predictors]) #make predictions on test set
    predictions = pd.DataFrame(predictions, columns=['predictions'], index=test.index) #name the column 'predictions'
    combination = pd.concat([test[['Player', 'Share']], predictions], axis =1) #Combine  'Player' and 'Share' values with predicted 'predictions' values
    all_predictions.append(combination) #append combination DataFrame to the list of all predictions
    aps.append(find_ap(combination)) #calculate and append the average precision for the current combination
```

```python
#average percision of model

sum(aps) / len(aps)
```

```python
def add_ranks(combination):
    combination = combination.sort_values('Share', ascending=False) #sort by share of votes
    combination['Rk'] = list(range(1, combination.shape[0]+1)) #ranking
    combination = combination.sort_values('predictions', ascending=False)
    combination['Predicted_Rk'] = list(range(1, combination.shape[0]+1)) #predicted rank
    combination['Diff'] = combination['Rk'] - combination['Predicted_Rk'] #difference between ranks
    return combination
```

```python
combination = combination.sort_values('Share', ascending=False) #sort by share of votes
combination['Rk'] = list(range(1, combination.shape[0]+1)) #ranking
combination = combination.sort_values('predictions', ascending=False)
combination['Predicted_Rk'] = list(range(1, combination.shape[0]+1)) #predicted rank
combination['Diff'] = combination['Rk'] - combination['Predicted_Rk'] #difference between ranks
```

```python
def backtest (mvp,model,year, predictors):
    aps=[]
    all_predictions = []
    for year in years[3:]:
        train = mvp[mvp['Year'] < year]  #training set with data from years less than the current year
        test= mvp[mvp['Year'] == year] #create test set with data from the current year
        reg.fit(train[predictors], train['Share']) #fit regression model using training data
        predictions = reg.predict(test[predictors]) #make predictions on test set
        predictions = pd.DataFrame(predictions, columns=['predictions'], index=test.index) #name the column 'predictions'
        combination = pd.concat([test[['Player', 'Share']], predictions], axis =1) #Combine  'Player' and 'Share' values with predicted 'predictions' values
        combination = add_ranks(combination)
        all_predictions.append(combination) #append combination DataFrame to the list of all predictions
        aps.append(find_ap(combination)) #calculate and append the average precision for the current combination
    return sum(aps)/len(aps), aps, pd.concat(all_predictions)
```

```python
mean_ap, aps, all_predictions = backtest(mvp, reg, years[2:], predictors)
```
