```python
import pandas as pd
import requests
pd.set_option('display.max_columns', None)  # so we can see all columns in a wide dataframe.
import time
import numpy as np
```

```python
years = list(range(2015,2023))
```

```python
url_start = 'https://www.basketball-reference.com/awards/awards_{}.html'
```

```python
# download mvp votes with requests

for year in years:
  url = url_start.format(year)
  data = requests.get(url)

  with open("mvp/{}.html".format(year), "w+") as f:
    f.write(data.text)
```

# **Parsing the Votes table with Beautifulsoup**


```python
# Import BeautifulSoup class from the bs4 module

from bs4 import BeautifulSoup
```

```python
# parse & open a single year

with open("mvp/2023.html") as f:
  page = f.read()
```

```python
# creates a parser class that allows you to extract table from page

soup = BeautifulSoup(page, "html.parser")
```

```python
# delete overhead html heading

soup.find('tr', class_="over_header").decompose()
```

```python
#find specific table based on id

mvp_table = soup.find(id="mvp")
mvp_2023 = pd.read_html(str(mvp_table))[0]
```

```python
mvp_2023
```

```python
mvp_2023['Year'] = '2023'
mvp_2023
```

```python
# create mvp table

dfs= []
for year in years:
  with open("mvp/{}.html".format(year)) as f:
    page = f.read()
  soup = BeautifulSoup(page, "html.parser")
  soup.find('tr', class_="over_header").decompose()
  mvp_table = soup.find(id="mvp") #locates table based on mvp id
  mvp = pd.read_html(str(mvp_table))[0] # reads in html
  mvp['Year'] = year # creates year column so we know which year data came from

  dfs.append(mvp)
```

```python
# combine dataframes

mvps = pd.concat(dfs)
mvps
```

```python
#add 2023 mvp votes

mvps = pd.concat([mvps,mvp_2023])
mvps
```

```python
mvps.to_csv("mvps.csv")
```

# **Downloading Team Data**

```python
team_standing_url = 'https://www.basketball-reference.com/leagues/NBA_{}_standings.html'
```

```python
# division standings

for year in years:
    url = team_standing_url.format(year)

    data = requests.get(url)

    with open("team/{}.html".format(year), "w+") as f:
         f.write(data.text)
```

# **Parse Team data with Beautifulsoup**

```python
#open team file

dfs = []
for year in years:
    with open("team/{}.html".format(year)) as f:
         page = f.read()

    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="thread")#.decompose() #removes headers
    team_table = soup.find(id="divs_standings_E")
    team = pd.read_html(str(team_table))[0]
    team['Year'] = year # creates year column so we know which year data came from
    team['Team'] = team["Eastern Conference"]
    del team["Eastern Conference"]
    dfs.append(team)

    soup = BeautifulSoup(page,"html.parser")
    soup.find('tr', class_="thread")#.decompose() #removes headers
    team_table = soup.find(id="divs_standings_W")
    team = pd.read_html(str(team_table))[0]
    team['Year'] = year # creates year column so we know which year data came from
    team['Team'] = team["Western Conference"]
    del team["Western Conference"]
    dfs.append(team)
```

```python
teams = pd.concat(dfs)
teams
```

```python
teams.to_csv("teams.csv")
```
